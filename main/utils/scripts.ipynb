```python
subjects = df.columns.tolist()
subjects
```

```python
from datetime import datetime

data = {}
for i,r in df.iterrows():
    # parse date of the format mm/dd/yyyy into datetime object and then convert it to yyyy-mm-dd
    d = datetime.strptime(i, '%m/%d/%Y')
    date = d.strftime('%Y-%m-%d')
    # print(date)
    data[date] = {
        j:{
            "body_mass":str(k),
            "weight_lifted":0,
        }
        for j,k in zip(subjects, r)
    }
    
data
```

```python
data['meta'] = {
    'subjects':subjects
}
data
```
